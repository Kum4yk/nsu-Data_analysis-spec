Предельная ошибка выборки:
$$ \Delta_{\overline{X}} = t \cdot \sqrt{\frac{s^2 \cdot (1 - \frac{n}{N})}{n}} $$

- n - объём выборки
- N - объём генеральной совокупности
- s$^2$ - несмещённая дисперсия
- t - коэффициент доверия Стьюдента

p = 95%, t = 1.96

$$ n = \frac{t^2 \cdot s^2 \cdot N}{ \Delta_{\overline{X}}^2 \cdot N + t^2 \cdot s^2} $$

```r
sample_volume <- function(t, s, N, Delta) {
    return((t*t*s*N)/(Delta*Delta*N + t*t*s))
} 
```
