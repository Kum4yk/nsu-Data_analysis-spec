2) Имеются следующие данные по количеству лотерейных билетов, проданных за день в киосках города N: 79, 85, 93, 99, 112, 116, 121, 127, 138, 141, 154, 169, 173, 188, 192, 213, 256, 274, 299, 303.

Сколько билетов будет продано за день с вероятностью 30%?
- Не менее 188 (+)
- Не менее 116
- Не более 116
- Не более 188 

```r
percent = 30
data = c(79, 85, 93, 99, 112, 116, 121, 127, 138, 141, 154, 169, 173, 188, 192, 213, 256, 274, 299, 303)
data[ceiling((100 - percent)/100*length(data))]
```

8) Имеется следующее распределение времени (в часах), затрачиваемого студентами на прохождение онлайн-курса: 1 15 20 22 27 31 34 37 39 40 42 43 46 49 50 51 59 60 60 115 138.

Чему равна медиана распределения?

```r
data = c(1, 15, 20, 22, 27, 31, 34, 37, 39, 40, 42, 43, 46, 49, 50, 51, 59, 60, 60, 115, 138)
print(c(median(data)))
```
